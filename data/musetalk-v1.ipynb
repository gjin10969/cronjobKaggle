```python
!sh /kaggle/working/MuseTalk/download_weights.sh
from huggingface_hub import hf_hub_download
import gdown
import os

CheckpointsDir = "models"

os.makedirs("models/musetalk", exist_ok=True)
os.makedirs("models/musetalkV15", exist_ok=True)
os.makedirs("models/syncnet", exist_ok=True)
os.makedirs("models/dwpose", exist_ok=True)
os.makedirs("models/face-parse-bisent", exist_ok=True)
os.makedirs("models/sd-vae", exist_ok=True)
os.makedirs("models/whisper", exist_ok=True)

# MuseTalk V1.0
hf_hub_download("TMElyralab/MuseTalk", "musetalk/musetalk.json", local_dir=CheckpointsDir)
hf_hub_download("TMElyralab/MuseTalk", "musetalk/pytorch_model.bin", local_dir=CheckpointsDir)

# MuseTalk V1.5
hf_hub_download("TMElyralab/MuseTalk", "musetalkV15/musetalk.json", local_dir=CheckpointsDir)
hf_hub_download("TMElyralab/MuseTalk", "musetalkV15/unet.pth", local_dir=CheckpointsDir)

# SD-VAE
hf_hub_download("stabilityai/sd-vae-ft-mse", "config.json", local_dir=f"{CheckpointsDir}/sd-vae")
hf_hub_download("stabilityai/sd-vae-ft-mse", "diffusion_pytorch_model.bin", local_dir=f"{CheckpointsDir}/sd-vae")

# Whisper tiny
hf_hub_download("openai/whisper-tiny", "config.json", local_dir=f"{CheckpointsDir}/whisper")
hf_hub_download("openai/whisper-tiny", "pytorch_model.bin", local_dir=f"{CheckpointsDir}/whisper")
hf_hub_download("openai/whisper-tiny", "preprocessor_config.json", local_dir=f"{CheckpointsDir}/whisper")

# DWPose
hf_hub_download("yzd-v/DWPose", "dw-ll_ucoco_384.pth", local_dir=f"{CheckpointsDir}/dwpose")

# SyncNet (LatentSync)
hf_hub_download("ByteDance/LatentSync", "latentsync_syncnet.pt", local_dir=f"{CheckpointsDir}/syncnet")

# Face parse bisent (Google Drive)
gdown.download("https://drive.google.com/uc?id=154JgKpzCPW82qINcVieuPH3fZ2e0P812",
               f"{CheckpointsDir}/face-parse-bisent/79999_iter.pth",
               quiet=False)

# ResNet18
os.system(f"curl -L https://download.pytorch.org/models/resnet18-5c106cde.pth -o {CheckpointsDir}/face-parse-bisent/resnet18-5c106cde.pth")

print("✅ All weights downloaded successfully!")
```
